**Two-Pass Methodology**

**Step 1 (Time-Series Regressions)**  
For each portfolio p:

Rp_t - Rf_t = α_p + β_p' f_t + ε_p_t

Estimate β_p for each portfolio.

**Step 2 (Cross-Sectional Regressions)**  
For each month t:

R_p,t - R_f,t = λ_0t + λ_t' β_p + η_p,t

Compute average λ̄ and their standard errors (Fama–MacBeth).

**Shanken Correction (1992):**

Var(λ̂) = (1 + μ_f' Σ_f^{-1} μ_f) * Var_FM(λ̂)

where μ_f and Σ_f are mean and covariance of factor returns.

This adjusts for sampling error in β estimates.

## Perform time-series regressions

### Subtask:
For each portfolio, regress the excess returns on the market excess returns to get the beta estimates.


**Reasoning**:
I will identify the portfolio excess return columns, then iterate through them to perform time-series regressions of each portfolio's excess return on the market excess return, storing the estimated betas.



```python
import statsmodels.api as sm

portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

portfolio_betas = {}

X = sm.add_constant(merged_df['market_excess_return'])

for portfolio_col in portfolio_excess_return_cols:
    y = merged_df[portfolio_col]
    # Drop rows with NaN values in the current portfolio's excess return
    valid_indices = y.dropna().index
    y_cleaned = y.loc[valid_indices]
    X_cleaned = X.loc[valid_indices]

    model = sm.OLS(y_cleaned, X_cleaned)
    results = model.fit()
    portfolio_betas[portfolio_col] = results.params['market_excess_return']

print("Estimated Betas for each portfolio:")
for portfolio, beta in portfolio_betas.items():
    print(f"{portfolio}: {beta:.4f}")
```

## Apply shanken correction

### Subtask:
Adjust the standard errors of the lambdas using the Shanken (1992) correction.


**Reasoning**:
Calculate the Shanken correction factor, apply it to the Fama-MacBeth standard errors, calculate the t-statistics, and print the corrected standard errors and t-statistics.



```python
import numpy as np

# 1. Calculate the mean vector and covariance matrix of the factor returns
# In this case, the only factor is market excess return.
mu_f = merged_df['market_excess_return'].mean()
# Corrected: Use var() for the variance of a single variable
Sigma_f = merged_df['market_excess_return'].var()

# 2. Calculate the Shanken correction factor
# The formula is (1 + mu_f' * inv(Sigma_f) * mu_f)
# For a single factor, this simplifies to (1 + mu_f^2 / Sigma_f)
shanken_correction_factor = (1 + (mu_f**2) / Sigma_f)

# 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
lambda_0_shanken_variance = (lambda_0_sem**2) * shanken_correction_factor
lambda_market_shanken_variance = (lambda_market_sem**2) * shanken_correction_factor

# 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
lambda_0_shanken_sem = np.sqrt(lambda_0_shanken_variance)
lambda_market_shanken_sem = np.sqrt(lambda_market_shanken_variance)

# 5. Calculate the t-statistics for the lambdas using the mean lambdas and the Shanken-corrected standard errors
lambda_0_shanken_tstat = lambda_0_mean / lambda_0_shanken_sem
lambda_market_shanken_tstat = lambda_market_mean / lambda_market_shanken_sem

# 6. Print the Shanken-corrected standard errors and t-statistics
print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem:.4f}")
print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat:.4f}")
print(f"Shanken-corrected Standard Error of lambda_market: {lambda_market_shanken_sem:.4f}")
print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat:.4f}")
```

# Model 2: HCAPM

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd # Ensure pandas is imported

# 1. Define human capital return: Rh_t = Δlog(Labor_Income_t)
# Using income_growth which is already calculated as pct_change, approximate with log(1+pct_change/100)
merged_df['human_capital_return'] = np.log(1 + merged_df['income_growth'] / 100)

# 2. Perform time-series regressions
print("Performing time-series regressions for HCAPM...")
portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

portfolio_betas_hcapm = {}

# Define independent variables: constant, market excess return, human capital return
X_ts = merged_df[['market_excess_return', 'human_capital_return']].copy()
X_ts = sm.add_constant(X_ts)

for portfolio_col in portfolio_excess_return_cols:
    y_ts = merged_df[portfolio_col]
    # Drop rows with NaN values in the current portfolio's excess return and corresponding rows from X_ts
    valid_indices = y_ts.dropna().index
    y_ts_cleaned = y_ts.loc[valid_indices]
    X_ts_cleaned = X_ts.loc[valid_indices]

    # Drop rows with NaN values in X_ts_cleaned before fitting the model
    valid_X_indices = X_ts_cleaned.dropna().index
    y_ts_cleaned = y_ts_cleaned.loc[valid_X_indices]
    X_ts_cleaned = X_ts_cleaned.loc[valid_X_indices]

    if not X_ts_cleaned.empty and not y_ts_cleaned.empty:
        model_ts = sm.OLS(y_ts_cleaned, X_ts_cleaned)
        results_ts = model_ts.fit()
        # Store the estimated betas
        portfolio_betas_hcapm[portfolio_col] = {
            'beta_market': results_ts.params['market_excess_return'],
            'beta_human_capital': results_ts.params['human_capital_return']
        }
    else:
         # Handle cases where there are no valid data points for regression
        portfolio_betas_hcapm[portfolio_col] = {
            'beta_market': np.nan,
            'beta_human_capital': np.nan
        }

print("Time-series regressions complete. Estimated Betas for HCAPM:")
for portfolio, betas in portfolio_betas_hcapm.items():
    print(f"{portfolio}:")
    print(f"  Beta Market: {betas['beta_market']:.4f}")
    print(f"  Beta Human Capital: {betas['beta_human_capital']:.4f}")


# 3. Perform cross-sectional regressions
print("\nPerforming cross-sectional regressions...")
lambda_estimates_hcapm = []

# Create a DataFrame of the estimated betas for cross-sectional regression
betas_df_hcapm = pd.DataFrame.from_dict(portfolio_betas_hcapm, orient='index')
betas_df_hcapm = sm.add_constant(betas_df_hcapm) # Add constant for lambda_0

for index, row in merged_df.iterrows():
    y_cs = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs and get the corresponding betas from betas_df_hcapm
    valid_portfolios = y_cs.dropna().index
    y_cs_cleaned = y_cs.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs = betas_df_hcapm.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 3 for intercept and two betas)
    if len(valid_portfolios) >= 3 and not betas_for_cs.empty:
        try:
            model_cs = sm.OLS(y_cs_cleaned, betas_for_cs)
            results_cs = model_cs.fit()
            # Ensure the number of parameters is 3 before appending (lambda_0, lambda_market, lambda_human_capital)
            if len(results_cs.params) == 3:
                lambda_estimates_hcapm.append(results_cs.params.tolist())
            else:
                 # Append NaNs if the number of parameters is not as expected
                lambda_estimates_hcapm.append([np.nan, np.nan, np.nan])
        except Exception as e:
            # Handle potential errors during regression
            print(f"Regression error at index {index}: {e}")
            lambda_estimates_hcapm.append([np.nan, np.nan, np.nan])
    else:
        # Append NaNs if regression cannot be performed
        lambda_estimates_hcapm.append([np.nan, np.nan, np.nan])

lambda_df_hcapm = pd.DataFrame(lambda_estimates_hcapm, columns=['lambda_0', 'lambda_market', 'lambda_human_capital'])
print("Cross-sectional regressions complete.")
#display(lambda_df_hcapm.head()) # Optional: display head of lambda_df_hcapm


# 4. Compute Fama–MacBeth lambdas and standard errors
print("\nComputing Fama-MacBeth lambdas and standard errors...")
lambda_0_mean_hcapm = lambda_df_hcapm['lambda_0'].mean()
lambda_0_sem_hcapm = lambda_df_hcapm['lambda_0'].sem()
lambda_market_mean_hcapm = lambda_df_hcapm['lambda_market'].mean()
lambda_market_sem_hcapm = lambda_df_hcapm['lambda_market'].sem()
lambda_human_capital_mean_hcapm = lambda_df_hcapm['lambda_human_capital'].mean()
lambda_human_capital_sem_hcapm = lambda_df_hcapm['lambda_human_capital'].sem()

print("Fama-MacBeth Lambda Estimates (HCAPM):")
print(f"Mean of lambda_0: {lambda_0_mean_hcapm:.4f}")
print(f"Standard Error of lambda_0: {lambda_0_sem_hcapm:.4f}")
print(f"Mean of lambda_market: {lambda_market_mean_hcapm:.4f}")
print(f"Standard Error of lambda_market: {lambda_market_sem_hcapm:.4f}")
print(f"Mean of lambda_human_capital: {lambda_human_capital_mean_hcapm:.4f}")
print(f"Standard Error of lambda_human_capital: {lambda_human_capital_sem_hcapm:.4f}")


# 5. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factors
# Factors are market excess return and human capital return
factors_hcapm_shanken = merged_df[['market_excess_return', 'human_capital_return']].dropna()
mu_f_hcapm = factors_hcapm_shanken.mean().values
Sigma_f_hcapm = factors_hcapm_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
# Use np.linalg.inv for matrix inverse and .T for transpose
try:
    Sigma_f_inv_hcapm = np.linalg.inv(Sigma_f_hcapm)
    shanken_correction_factor_hcapm = (1 + mu_f_hcapm.T @ Sigma_f_inv_hcapm @ mu_f_hcapm)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_hcapm = np.nan


if not np.isnan(shanken_correction_factor_hcapm):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_hcapm = (lambda_0_sem_hcapm**2) * shanken_correction_factor_hcapm
    lambda_market_shanken_variance_hcapm = (lambda_market_sem_hcapm**2) * shanken_correction_factor_hcapm
    lambda_human_capital_shanken_variance_hcapm = (lambda_human_capital_sem_hcapm**2) * shanken_correction_factor_hcapm

    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_hcapm = np.sqrt(lambda_0_shanken_variance_hcapm)
    lambda_market_shanken_sem_hcapm = np.sqrt(lambda_market_shanken_variance_hcapm)
    lambda_human_capital_shanken_sem_hcapm = np.sqrt(lambda_human_capital_shanken_variance_hcapm)

    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_hcapm = lambda_0_mean_hcapm / lambda_0_shanken_sem_hcapm
    lambda_market_shanken_tstat_hcapm = lambda_market_mean_hcapm / lambda_market_shanken_sem_hcapm
    lambda_human_capital_shanken_tstat_hcapm = lambda_human_capital_mean_hcapm / lambda_human_capital_shanken_sem_hcapm

    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (HCAPM):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_hcapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_market: {lambda_market_shanken_sem_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat_hcapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_human_capital: {lambda_human_capital_shanken_sem_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_human_capital: {lambda_human_capital_shanken_tstat_hcapm:.4f}")

else:
    print("\nShanken correction could not be applied.")

# 6. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_hcapm = []

for index, row in merged_df.iterrows():
    y_cs = row[portfolio_excess_return_cols]

    valid_portfolios = y_cs.dropna().index
    y_cs_cleaned = y_cs.loc[valid_portfolios].astype(float)

    betas_for_cs = betas_df_hcapm.loc[valid_portfolios].astype(float)

    if len(valid_portfolios) >= 3 and not betas_for_cs.empty:
        try:
            model_cs = sm.OLS(y_cs_cleaned, betas_for_cs)
            results_cs = model_cs.fit()
            if len(results_cs.params) == 3:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned.mean()
                ss_total = ((y_cs_cleaned - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs.predict(betas_for_cs)
                ss_residual = ((y_cs_cleaned - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_hcapm.append(cs_r_squared)
                else:
                    cs_r_squared_values_hcapm.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_hcapm.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_hcapm.append(np.nan)
    else:
        cs_r_squared_values_hcapm.append(np.nan)

average_cs_r_squared_hcapm = np.nanmean(cs_r_squared_values_hcapm)

print(f"Average Cross-Sectional R-squared (HCAPM): {average_cs_r_squared_hcapm:.4f}")


# 7. Report results (Summary)
print("\n--- HCAPM Model Results ---")
print(f"Mean lambda_market: {lambda_market_mean_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat_hcapm:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_market: N/A (Shanken correction not applied)")

print(f"Mean lambda_human_capital: {lambda_human_capital_mean_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_human_capital: {lambda_human_capital_shanken_tstat_hcapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_human_capital: N/A (Shanken correction not applied)")

print(f"Mean lambda_0: {lambda_0_mean_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_hcapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")

print(f"Average cross-sectional R-squared: {average_cs_r_squared_hcapm:.4f}")

print("\n--- HCAPM Summary ---")
print("Key Findings:")
print(f"* The estimated betas for the six portfolios show varying sensitivities to market and human capital risk.")
print(f"* The mean estimated market risk premium (lambda_market) is {lambda_market_mean_hcapm:.4f}.")
print(f"* The mean estimated human capital risk premium (lambda_human_capital) is {lambda_human_capital_mean_hcapm:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_hcapm:.4f}.")

if not np.isnan(shanken_correction_factor_hcapm):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - Market Risk Premium t-statistic: {lambda_market_shanken_tstat_hcapm:.4f}")
    print(f"  - Human Capital Risk Premium t-statistic: {lambda_human_capital_shanken_tstat_hcapm:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_hcapm:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_hcapm:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Compare the results of the HCAPM model to the CAPM model (previously estimated).")
print("- Estimate the CCAPM model and compare its results to the HCAPM and CAPM models.")
print("- Estimate the conditional models with cay as a factor and multiplier.")
print("- Estimate the Fama-French 3-Factor Model.")
print("- Summarize all model results in a comprehensive table.")
```

```python
# 4. Compute Fama–MacBeth standard errors for lambdas
print("\nComputing Fama-MacBeth standard errors...")
lambda_0_sem_ccapm = lambda_df_ccapm['lambda_0'].sem()
lambda_consumption_sem_ccapm = lambda_df_ccapm['lambda_consumption'].sem()

print("Fama-MacBeth Standard Errors (CCAPM):")
print(f"Standard Error of lambda_0: {lambda_0_sem_ccapm:.4f}")
print(f"Standard Error of lambda_consumption: {lambda_consumption_sem_ccapm:.4f}")


# 5. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factor returns
# In this case, the only factor is consumption growth.
factors_ccapm_shanken = merged_df[['consumption_growth']].dropna()
mu_f_ccapm = factors_ccapm_shanken.mean().values
Sigma_f_ccapm = factors_ccapm_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
# Use np.linalg.inv for matrix inverse and .T for transpose
try:
    Sigma_f_inv_ccapm = np.linalg.inv(Sigma_f_ccapm)
    shanken_correction_factor_ccapm = (1 + mu_f_ccapm.T @ Sigma_f_inv_ccapm @ mu_f_ccapm)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_ccapm = np.nan


if not np.isnan(shanken_correction_factor_ccapm):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_ccapm = (lambda_0_sem_ccapm**2) * shanken_correction_factor_ccapm
    lambda_consumption_shanken_variance_ccapm = (lambda_consumption_sem_ccapm**2) * shanken_correction_factor_ccapm

    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_ccapm = np.sqrt(lambda_0_shanken_variance_ccapm)
    lambda_consumption_shanken_sem_ccapm = np.sqrt(lambda_consumption_shanken_variance_ccapm)

    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_ccapm = lambda_0_mean_ccapm / lambda_0_shanken_sem_ccapm
    lambda_consumption_shanken_tstat_ccapm = lambda_consumption_mean_ccapm / lambda_consumption_shanken_sem_ccapm

    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (CCAPM):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_ccapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_ccapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_consumption: {lambda_consumption_shanken_sem_ccapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_consumption: {lambda_consumption_shanken_tstat_ccapm:.4f}")

else:
    print("\nShanken correction could not be applied.")


# 6. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_ccapm = []

for index, row in merged_df.iterrows():
    y_cs_ccapm = row[portfolio_excess_return_cols]

    valid_portfolios = y_cs_ccapm.dropna().index
    y_cs_cleaned_ccapm = y_cs_ccapm.loc[valid_portfolios].astype(float)

    gammas_for_cs_ccapm = gammas_df_ccapm.loc[valid_portfolios].astype(float)

    if len(valid_portfolios) >= 2 and not gammas_for_cs_ccapm.empty:
        try:
            model_cs_ccapm = sm.OLS(y_cs_cleaned_ccapm, gammas_for_cs_ccapm)
            results_cs_ccapm = model_cs_ccapm.fit()
            if len(results_cs_ccapm.params) == 2:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned_ccapm.mean()
                ss_total = ((y_cs_cleaned_ccapm - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs_ccapm.predict(gammas_for_cs_ccapm)
                ss_residual = ((y_cs_cleaned_ccapm - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_ccapm.append(cs_r_squared)
                else:
                    cs_r_squared_values_ccapm.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_ccapm.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_ccapm.append(np.nan)
    else:
        cs_r_squared_values_ccapm.append(np.nan)

average_cs_r_squared_ccapm = np.nanmean(cs_r_squared_values_ccapm)

print(f"Average Cross-Sectional R-squared (CCAPM): {average_cs_r_squared_ccapm:.4f}")

# 7. Report results (Summary)
print("\n--- CCAPM Model Results ---")
print(f"Mean lambda_c (consumption): {lambda_consumption_mean_ccapm:.4f}")
if not np.isnan(shanken_correction_factor_ccapm):
    print(f"Shanken-corrected t-statistic for lambda_c: {lambda_consumption_shanken_tstat_ccapm:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_c: N/A (Shanken correction not applied)")

print(f"Mean lambda_0: {lambda_0_mean_ccapm:.4f}")
if not np.isnan(shanken_correction_factor_ccapm):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_ccapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")

print(f"Average cross-sectional R-squared: {average_cs_r_squared_ccapm:.4f}")

print("\n--- CCAPM Summary ---")
print("Key Findings:")
print(f"* The estimated gammas for the six portfolios show varying sensitivities to consumption growth.")
print(f"* The mean estimated consumption risk premium (lambda_c) is {lambda_consumption_mean_ccapm:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_ccapm:.4f}.")

if not np.isnan(shanken_correction_factor_ccapm):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - Consumption Risk Premium t-statistic: {lambda_consumption_shanken_tstat_ccapm:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_ccapm:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_ccapm:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Compare the results of the CCAPM model to the CAPM and HCAPM models.")
print("- Estimate the conditional models with cay as a factor and multiplier.")
print("- Estimate the Fama-French 3-Factor Model.")
print("- Summarize all model results in a comprehensive table.")
```

```python
import statsmodels.api as sm
import numpy as np

# 1. Define a list containing the names of the portfolio excess return columns.
portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

# 2. Initialize an empty dictionary to store the estimated betas for each portfolio.
portfolio_betas_capm_cay_interaction = {}

# 3. Define the independent variables for the time-series regression as the 'market_excess_return', 'cay_standardized', and 'market_excess_return_cay_interaction' columns from merged_df.
X_ts = merged_df[['market_excess_return', 'cay_standardized', 'market_excess_return_cay_interaction']].copy()

# 4. Add a constant term to the independent variables DataFrame using sm.add_constant.
X_ts = sm.add_constant(X_ts)

# 5. Iterate through each portfolio excess return column:
print("Performing time-series regressions for CAPM + cay + Interaction...")
for portfolio_col in portfolio_excess_return_cols:
    # a. Define the dependent variable as the current portfolio's excess return.
    y_ts = merged_df[portfolio_col]

    # b. Drop rows with NaN values in the dependent variable and the corresponding rows in the independent variables DataFrame.
    valid_indices = y_ts.dropna().index
    y_ts_cleaned = y_ts.loc[valid_indices]
    X_ts_cleaned = X_ts.loc[valid_indices]

    # Drop rows with NaN values in X_ts_cleaned before fitting the model
    valid_X_indices = X_ts_cleaned.dropna().index
    y_ts_cleaned = y_ts_cleaned.loc[valid_X_indices]
    X_ts_cleaned = X_ts_cleaned.loc[valid_X_indices]

    if not X_ts_cleaned.empty and not y_ts_cleaned.empty:
        # c. Fit an OLS model using sm.OLS with the cleaned dependent and independent variables.
        model_ts = sm.OLS(y_ts_cleaned, X_ts_cleaned)
        results_ts = model_ts.fit()

        # d. Store the estimated coefficients for 'market_excess_return', 'cay_standardized', and 'market_excess_return_cay_interaction' in the dictionary created in step 2, using the portfolio column name as the key.
        portfolio_betas_capm_cay_interaction[portfolio_col] = {
            'beta_market': results_ts.params['market_excess_return'],
            'beta_cay': results_ts.params['cay_standardized'],
            'beta_market_cay_interaction': results_ts.params['market_excess_return_cay_interaction']
        }
    else:
         # Handle cases where there are no valid data points for regression
        portfolio_betas_capm_cay_interaction[portfolio_col] = {
            'beta_market': np.nan,
            'beta_cay': np.nan,
            'beta_market_cay_interaction': np.nan
        }

# 6. Print the estimated betas for each portfolio.
print("Time-series regressions complete. Estimated Betas for CAPM + cay + Interaction:")
for portfolio, betas in portfolio_betas_capm_cay_interaction.items():
    print(f"{portfolio}:")
    print(f"  Beta Market: {betas['beta_market']:.4f}")
    print(f"  Beta CAY: {betas['beta_cay']:.4f}")
    print(f"  Beta Market x CAY Interaction: {betas['beta_market_cay_interaction']:.4f}")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Create a DataFrame of the estimated betas for cross-sectional regression
betas_df_ccapm_cay = pd.DataFrame.from_dict(portfolio_betas_ccapm_cay, orient='index')
betas_df_ccapm_cay = sm.add_constant(betas_df_ccapm_cay) # Add constant for lambda_0

# Perform cross-sectional regressions
print("\nPerforming cross-sectional regressions for CCAPM + cay...")
lambda_estimates_ccapm_cay = []

portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

for index, row in merged_df.iterrows():
    y_cs = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs and get the corresponding betas from betas_df_ccapm_cay
    valid_portfolios = y_cs.dropna().index
    y_cs_cleaned = y_cs.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs = betas_df_ccapm_cay.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 3 for intercept and two betas)
    if len(valid_portfolios) >= 3 and not betas_for_cs.empty:
        try:
            model_cs = sm.OLS(y_cs_cleaned, betas_for_cs)
            results_cs = model_cs.fit()
            # Ensure the number of parameters is 3 before appending (lambda_0, lambda_consumption, lambda_cay)
            if len(results_cs.params) == 3:
                lambda_estimates_ccapm_cay.append(results_cs.params.tolist())
            else:
                 # Append NaNs if the number of parameters is not as expected
                lambda_estimates_ccapm_cay.append([np.nan, np.nan, np.nan])
        except Exception as e:
            # Handle potential errors during regression
            print(f"Regression error at index {index}: {e}")
            lambda_estimates_ccapm_cay.append([np.nan, np.nan, np.nan])
    else:
        # Append NaNs if regression cannot be performed
        lambda_estimates_ccapm_cay.append([np.nan, np.nan, np.nan])

lambda_df_ccapm_cay = pd.DataFrame(lambda_estimates_ccapm_cay, columns=['lambda_0', 'lambda_consumption', 'lambda_cay'])
print("Cross-sectional regressions complete.")
#display(lambda_df_ccapm_cay.head()) # Optional: display head of lambda_df_ccapm_cay

# 3. Compute Fama–MacBeth lambdas and standard errors
print("\nComputing Fama-MacBeth lambdas and standard errors for CCAPM + cay...")
lambda_0_mean_ccapm_cay = lambda_df_ccapm_cay['lambda_0'].mean()
lambda_0_sem_ccapm_cay = lambda_df_ccapm_cay['lambda_0'].sem()
lambda_consumption_mean_ccapm_cay = lambda_df_ccapm_cay['lambda_consumption'].mean()
lambda_consumption_sem_ccapm_cay = lambda_df_ccapm_cay['lambda_consumption'].sem()
lambda_cay_mean_ccapm_cay = lambda_df_ccapm_cay['lambda_cay'].mean()
lambda_cay_sem_ccapm_cay = lambda_df_ccapm_cay['lambda_cay'].sem()


print("Fama-MacBeth Lambda Estimates (CCAPM + cay):")
print(f"Mean of lambda_0: {lambda_0_mean_ccapm_cay:.4f}")
print(f"Standard Error of lambda_0: {lambda_0_sem_ccapm_cay:.4f}")
print(f"Mean of lambda_consumption: {lambda_consumption_mean_ccapm_cay:.4f}")
print(f"Standard Error of lambda_consumption: {lambda_consumption_sem_ccapm_cay:.4f}")
print(f"Mean of lambda_cay: {lambda_cay_mean_ccapm_cay:.4f}")
print(f"Standard Error of lambda_cay: {lambda_cay_sem_ccapm_cay:.4f}")


# 4. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factors
# Factors are consumption growth and standardized cay
factors_ccapm_cay_shanken = merged_df[['consumption_growth', 'cay_standardized']].dropna()
mu_f_ccapm_cay = factors_ccapm_cay_shanken.mean().values
Sigma_f_ccapm_cay = factors_ccapm_cay_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
try:
    Sigma_f_inv_ccapm_cay = np.linalg.inv(Sigma_f_ccapm_cay)
    shanken_correction_factor_ccapm_cay = (1 + mu_f_ccapm_cay.T @ Sigma_f_inv_ccapm_cay @ mu_f_ccapm_cay)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_ccapm_cay = np.nan


if not np.isnan(shanken_correction_factor_ccapm_cay):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_ccapm_cay = (lambda_0_sem_ccapm_cay**2) * shanken_correction_factor_ccapm_cay
    lambda_consumption_shanken_variance_ccapm_cay = (lambda_consumption_sem_ccapm_cay**2) * shanken_correction_factor_ccapm_cay
    lambda_cay_shanken_variance_ccapm_cay = (lambda_cay_sem_ccapm_cay**2) * shanken_correction_factor_ccapm_cay


    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_ccapm_cay = np.sqrt(lambda_0_shanken_variance_ccapm_cay)
    lambda_consumption_shanken_sem_ccapm_cay = np.sqrt(lambda_consumption_shanken_variance_ccapm_cay)
    lambda_cay_shanken_sem_ccapm_cay = np.sqrt(lambda_cay_shanken_variance_ccapm_cay)


    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_ccapm_cay = lambda_0_mean_ccapm_cay / lambda_0_shanken_sem_ccapm_cay
    lambda_consumption_shanken_tstat_ccapm_cay = lambda_consumption_mean_ccapm_cay / lambda_consumption_shanken_sem_ccapm_cay
    lambda_cay_shanken_tstat_ccapm_cay = lambda_cay_mean_ccapm_cay / lambda_cay_shanken_sem_ccapm_cay


    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (CCAPM + cay):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_ccapm_cay:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_ccapm_cay:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_consumption: {lambda_consumption_shanken_sem_ccapm_cay:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_consumption: {lambda_consumption_shanken_tstat_ccapm_cay:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_cay: {lambda_cay_shanken_sem_ccapm_cay:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_cay: {lambda_cay_shanken_tstat_ccapm_cay:.4f}")

else:
    print("\nShanken correction could not be applied.")

# 5. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_ccapm_cay = []

for index, row in merged_df.iterrows():
    y_cs = row[portfolio_excess_return_cols]

    valid_portfolios = y_cs.dropna().index
    y_cs_cleaned = y_cs.loc[valid_portfolios].astype(float)

    betas_for_cs = betas_df_ccapm_cay.loc[valid_portfolios].astype(float)

    if len(valid_portfolios) >= 3 and not betas_for_cs.empty:
        try:
            model_cs = sm.OLS(y_cs_cleaned, betas_for_cs)
            results_cs = model_cs.fit()
            if len(results_cs.params) == 3:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned.mean()
                ss_total = ((y_cs_cleaned - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs.predict(betas_for_cs)
                ss_residual = ((y_cs_cleaned - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_ccapm_cay.append(cs_r_squared)
                else:
                    cs_r_squared_values_ccapm_cay.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_ccapm_cay.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_ccapm_cay.append(np.nan)
    else:
        cs_r_squared_values_ccapm_cay.append(np.nan)


average_cs_r_squared_ccapm_cay = np.nanmean(cs_r_squared_values_ccapm_cay)

print(f"Average Cross-Sectional R-squared (CCAPM + cay): {average_cs_r_squared_ccapm_cay:.4f}")


# 6. Report results (Summary)
print("\n--- CCAPM + cay Model Results ---")
print(f"Mean lambda_consumption: {lambda_consumption_mean_ccapm_cay:.4f}")
if not np.isnan(shanken_correction_factor_ccapm_cay):
    print(f"Shanken-corrected t-statistic for lambda_consumption: {lambda_consumption_shanken_tstat_ccapm_cay:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_consumption: N/A (Shanken correction not applied)")

print(f"Mean lambda_cay: {lambda_cay_mean_ccapm_cay:.4f}")
if not np.isnan(shanken_correction_factor_ccapm_cay):
    print(f"Shanken-corrected t-statistic for lambda_cay: {lambda_cay_shanken_tstat_ccapm_cay:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_cay: N/A (Shanken correction not applied)")

print(f"Mean lambda_0: {lambda_0_mean_ccapm_cay:.4f}")
if not np.isnan(shanken_correction_factor_ccapm_cay):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_ccapm_cay:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")


print(f"Average cross-sectional R-squared: {average_cs_r_squared_ccapm_cay:.4f}")

print("\n--- CCAPM + cay Summary ---")
print("Key Findings:")
print(f"* The estimated betas for the six portfolios show varying sensitivities to consumption growth and cay.")
print(f"* The mean estimated consumption risk premium (lambda_consumption) is {lambda_consumption_mean_ccapm_cay:.4f}.")
print(f"* The mean estimated cay risk premium (lambda_cay) is {lambda_cay_mean_ccapm_cay:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_ccapm_cay:.4f}.")

if not np.isnan(shanken_correction_factor_ccapm_cay):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - Consumption Risk Premium t-statistic: {lambda_consumption_shanken_tstat_ccapm_cay:.4f}")
    print(f"  - CAY Risk Premium t-statistic: {lambda_cay_shanken_tstat_ccapm_cay:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_ccapm_cay:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_ccapm_cay:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Summarize all model results in a comprehensive table.")
```

## Model 5

### Subtask:
Estimate conditional asset pricing models with `cay` only as multiplier.


**Reasoning**:
I need to estimate conditional asset pricing models with `cay` only as a multiplier. This involves creating interaction terms between the base model factors and the standardized `cay` variable, performing time-series regressions to get betas for these interaction terms, and then performing cross-sectional regressions to get the lambdas, followed by Shanken correction and R-squared calculation. I will start by creating the interaction terms for the CAPM, HCAPM, and CCAPM models with standardized `cay` as the multiplier.



```python
merged_df['market_excess_return_x_cay'] = merged_df['market_excess_return'] * merged_df['cay_standardized']
merged_df['human_capital_return_x_cay'] = merged_df['human_capital_return'] * merged_df['cay_standardized']
merged_df['consumption_growth_x_cay'] = merged_df['consumption_growth'] * merged_df['cay_standardized']

display(merged_df.head())
```

**Reasoning**:
I have created the necessary interaction terms. Now I will perform the time-series regressions for the conditional CAPM model with `cay` as a multiplier, which means regressing portfolio excess returns on the market excess return and the market excess return x cay interaction term. I will then proceed to perform the cross-sectional regressions, calculate Fama-MacBeth lambdas and standard errors, apply the Shanken correction, and compute the average cross-sectional R².



```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

# 1. Define a list containing the names of the portfolio excess return columns.
portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

# 2. Create a dictionary to store the estimated betas for each portfolio (Conditional CAPM).
portfolio_betas_conditional_capm = {}

# 3. Define the independent variables for the time-series regression as the 'market_excess_return' and 'market_excess_return_x_cay' columns.
X_ts_conditional_capm = merged_df[['market_excess_return', 'market_excess_return_x_cay']].copy()

# 4. Add a constant term to the independent variables DataFrame.
X_ts_conditional_capm = sm.add_constant(X_ts_conditional_capm)

# 5. Iterate through each portfolio excess return column:
print("Performing time-series regressions for Conditional CAPM (cay as multiplier)...")
for portfolio_col in portfolio_excess_return_cols:
    # a. Define the dependent variable as the current portfolio's excess return.
    y_ts_conditional_capm = merged_df[portfolio_col]

    # b. Drop rows with NaN values in the dependent variable and the corresponding rows in the independent variables DataFrame.
    valid_indices = y_ts_conditional_capm.dropna().index
    y_ts_cleaned_conditional_capm = y_ts_conditional_capm.loc[valid_indices]
    X_ts_cleaned_conditional_capm = X_ts_conditional_capm.loc[valid_indices]

    # Drop rows with NaN values in X_ts_cleaned_conditional_capm before fitting the model
    valid_X_indices = X_ts_cleaned_conditional_capm.dropna().index
    y_ts_cleaned_conditional_capm = y_ts_cleaned_conditional_capm.loc[valid_X_indices]
    X_ts_cleaned_conditional_capm = X_ts_cleaned_conditional_capm.loc[valid_X_indices]

    if not X_ts_cleaned_conditional_capm.empty and not y_ts_cleaned_conditional_capm.empty:
        # c. Fit an OLS model using sm.OLS with the cleaned dependent and independent variables.
        model_ts_conditional_capm = sm.OLS(y_ts_cleaned_conditional_capm, X_ts_cleaned_conditional_capm)
        results_ts_conditional_capm = model_ts_conditional_capm.fit()

        # d. Store the estimated coefficients in the dictionary.
        portfolio_betas_conditional_capm[portfolio_col] = {
            'beta_market': results_ts_conditional_capm.params['market_excess_return'],
            'beta_market_x_cay': results_ts_conditional_capm.params['market_excess_return_x_cay']
        }
    else:
         # Handle cases where there are no valid data points for regression
        portfolio_betas_conditional_capm[portfolio_col] = {
            'beta_market': np.nan,
            'beta_market_x_cay': np.nan
        }

# 6. Print the estimated betas for each portfolio.
print("Time-series regressions complete. Estimated Betas for Conditional CAPM:")
for portfolio, betas in portfolio_betas_conditional_capm.items():
    print(f"{portfolio}:")
    print(f"  Beta Market: {betas['beta_market']:.4f}")
    print(f"  Beta Market x CAY: {betas['beta_market_x_cay']:.4f}")


# 7. Create a DataFrame of the estimated betas for cross-sectional regression and add a constant for lambda_0.
betas_df_conditional_capm = pd.DataFrame.from_dict(portfolio_betas_conditional_capm, orient='index')
betas_df_conditional_capm = sm.add_constant(betas_df_conditional_capm) # Add constant for lambda_0

# 8. Perform cross-sectional regressions
print("\nPerforming cross-sectional regressions for Conditional CAPM...")
lambda_estimates_conditional_capm = []

for index, row in merged_df.iterrows():
    y_cs_conditional_capm = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs_conditional_capm and get the corresponding betas
    valid_portfolios = y_cs_conditional_capm.dropna().index
    y_cs_cleaned_conditional_capm = y_cs_conditional_capm.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs_conditional_capm = betas_df_conditional_capm.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 3 for intercept and two betas)
    if len(valid_portfolios) >= 3 and not betas_for_cs_conditional_capm.empty:
        try:
            model_cs_conditional_capm = sm.OLS(y_cs_cleaned_conditional_capm, betas_for_cs_conditional_capm)
            results_cs_conditional_capm = model_cs_conditional_capm.fit()
            # Ensure the number of parameters is 3 before appending (lambda_0, lambda_market, lambda_market_x_cay)
            if len(results_cs_conditional_capm.params) == 3:
                lambda_estimates_conditional_capm.append(results_cs_conditional_capm.params.tolist())
            else:
                 # Append NaNs if the number of parameters is not as expected
                lambda_estimates_conditional_capm.append([np.nan, np.nan, np.nan])
        except Exception as e:
            # Handle potential errors during regression
            print(f"Regression error at index {index}: {e}")
            lambda_estimates_conditional_capm.append([np.nan, np.nan, np.nan])
    else:
        # Append NaNs if regression cannot be performed
        lambda_estimates_conditional_capm.append([np.nan, np.nan, np.nan])

lambda_df_conditional_capm = pd.DataFrame(lambda_estimates_conditional_capm, columns=['lambda_0', 'lambda_market', 'lambda_market_x_cay'])
print("Cross-sectional regressions complete.")
#display(lambda_df_conditional_capm.head()) # Optional: display head of lambda_df_conditional_capm


# 9. Compute Fama–MacBeth lambdas and standard errors
print("\nComputing Fama-MacBeth lambdas and standard errors for Conditional CAPM...")
lambda_0_mean_conditional_capm = lambda_df_conditional_capm['lambda_0'].mean()
lambda_0_sem_conditional_capm = lambda_df_conditional_capm['lambda_0'].sem()
lambda_market_mean_conditional_capm = lambda_df_conditional_capm['lambda_market'].mean()
lambda_market_sem_conditional_capm = lambda_df_conditional_capm['lambda_market'].sem()
lambda_market_x_cay_mean_conditional_capm = lambda_df_conditional_capm['lambda_market_x_cay'].mean()
lambda_market_x_cay_sem_conditional_capm = lambda_df_conditional_capm['lambda_market_x_cay'].sem()


print("Fama-MacBeth Lambda Estimates (Conditional CAPM):")
print(f"Mean of lambda_0: {lambda_0_mean_conditional_capm:.4f}")
print(f"Standard Error of lambda_0: {lambda_0_sem_conditional_capm:.4f}")
print(f"Mean of lambda_market: {lambda_market_mean_conditional_capm:.4f}")
print(f"Standard Error of lambda_market: {lambda_market_sem_conditional_capm:.4f}")
print(f"Mean of lambda_market_x_cay: {lambda_market_x_cay_mean_conditional_capm:.4f}")
print(f"Standard Error of lambda_market_x_cay: {lambda_market_x_cay_sem_conditional_capm:.4f}")


# 10. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factors
# Factors are market excess return and market excess return x cay interaction
factors_conditional_capm_shanken = merged_df[['market_excess_return', 'market_excess_return_x_cay']].dropna()
mu_f_conditional_capm = factors_conditional_capm_shanken.mean().values
Sigma_f_conditional_capm = factors_conditional_capm_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
try:
    Sigma_f_inv_conditional_capm = np.linalg.inv(Sigma_f_conditional_capm)
    shanken_correction_factor_conditional_capm = (1 + mu_f_conditional_capm.T @ Sigma_f_inv_conditional_capm @ mu_f_conditional_capm)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_conditional_capm = np.nan


if not np.isnan(shanken_correction_factor_conditional_capm):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_conditional_capm = (lambda_0_sem_conditional_capm**2) * shanken_correction_factor_conditional_capm
    lambda_market_shanken_variance_conditional_capm = (lambda_market_sem_conditional_capm**2) * shanken_correction_factor_conditional_capm
    lambda_market_x_cay_shanken_variance_conditional_capm = (lambda_market_x_cay_sem_conditional_capm**2) * shanken_correction_factor_conditional_capm


    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_conditional_capm = np.sqrt(lambda_0_shanken_variance_conditional_capm)
    lambda_market_shanken_sem_conditional_capm = np.sqrt(lambda_market_shanken_variance_conditional_capm)
    lambda_market_x_cay_shanken_sem_conditional_capm = np.sqrt(lambda_market_x_cay_shanken_variance_conditional_capm)


    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_conditional_capm = lambda_0_mean_conditional_capm / lambda_0_shanken_sem_conditional_capm
    lambda_market_shanken_tstat_conditional_capm = lambda_market_mean_conditional_capm / lambda_market_shanken_sem_conditional_capm
    lambda_market_x_cay_shanken_tstat_conditional_capm = lambda_market_x_cay_mean_conditional_capm / lambda_market_x_cay_shanken_sem_conditional_capm


    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (Conditional CAPM):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_conditional_capm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_conditional_capm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_market: {lambda_market_shanken_sem_conditional_capm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat_conditional_capm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_market_x_cay: {lambda_market_x_cay_shanken_sem_conditional_capm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_market_x_cay: {lambda_market_x_cay_shanken_tstat_conditional_capm:.4f}")

else:
    print("\nShanken correction could not be applied.")


# 11. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_conditional_capm = []

for index, row in merged_df.iterrows():
    y_cs_conditional_capm = row[portfolio_excess_return_cols]

    valid_portfolios = y_cs_conditional_capm.dropna().index
    y_cs_cleaned_conditional_capm = y_cs_conditional_capm.loc[valid_portfolios].astype(float)

    betas_for_cs_conditional_capm = betas_df_conditional_capm.loc[valid_portfolios].astype(float)

    if len(valid_portfolios) >= 3 and not betas_for_cs_conditional_capm.empty:
        try:
            model_cs_conditional_capm = sm.OLS(y_cs_cleaned_conditional_capm, betas_for_cs_conditional_capm)
            results_cs_conditional_capm = model_cs_conditional_capm.fit()
            if len(results_cs_conditional_capm.params) == 3:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned_conditional_capm.mean()
                ss_total = ((y_cs_cleaned_conditional_capm - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs_conditional_capm.predict(betas_for_cs_conditional_capm)
                ss_residual = ((y_cs_cleaned_conditional_capm - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_conditional_capm.append(cs_r_squared)
                else:
                    cs_r_squared_values_conditional_capm.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_conditional_capm.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_conditional_capm.append(np.nan)
    else:
        cs_r_squared_values_conditional_capm.append(np.nan)


average_cs_r_squared_conditional_capm = np.nanmean(cs_r_squared_values_conditional_capm)

print(f"Average Cross-Sectional R-squared (Conditional CAPM): {average_cs_r_squared_conditional_capm:.4f}")


# 12. Report results (Summary)
print("\n--- Conditional CAPM Model Results (cay as multiplier) ---")
print(f"Mean lambda_market: {lambda_market_mean_conditional_capm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_capm):
    print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat_conditional_capm:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_market: N/A (Shanken correction not applied)")

print(f"Mean lambda_market_x_cay: {lambda_market_x_cay_mean_conditional_capm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_capm):
    print(f"Shanken-corrected t-statistic for lambda_market_x_cay: {lambda_market_x_cay_shanken_tstat_conditional_capm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_market_x_cay: N/A (Shanken correction not applied)")

print(f"Mean lambda_0: {lambda_0_mean_conditional_capm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_capm):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_conditional_capm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")

print(f"Average cross-sectional R-squared: {average_cs_r_squared_conditional_capm:.4f}")

print("\n--- Conditional CAPM Summary ---")
print("Key Findings:")
print(f"* The estimated betas for the six portfolios show varying sensitivities to market risk and conditional market risk (market x cay).")
print(f"* The mean estimated market risk premium (lambda_market) is {lambda_market_mean_conditional_capm:.4f}.")
print(f"* The mean estimated conditional market risk premium (lambda_market_x_cay) is {lambda_market_x_cay_mean_conditional_capm:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_conditional_capm:.4f}.")

if not np.isnan(shanken_correction_factor_conditional_capm):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - Market Risk Premium t-statistic: {lambda_market_shanken_tstat_conditional_capm:.4f}")
    print(f"  - Conditional Market Risk Premium t-statistic: {lambda_market_x_cay_shanken_tstat_conditional_capm:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_conditional_capm:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_conditional_capm:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Estimate the Conditional HCAPM model (cay as multiplier).")
print("- Estimate the Conditional CCAPM model (cay as multiplier).")
print("- Summarize all conditional model results and compare them to the unconditional models.")
```

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

# 1. Define a list containing the names of the portfolio excess return columns.
portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

# 2. Create a dictionary to store the estimated betas for each portfolio (Conditional HCAPM).
portfolio_betas_conditional_hcapm = {}

# 3. Define the independent variables for the time-series regression as 'market_excess_return', 'human_capital_return', and 'human_capital_return_x_cay'.
X_ts_conditional_hcapm = merged_df[['market_excess_return', 'human_capital_return', 'human_capital_return_x_cay']].copy()

# 4. Add a constant term to the independent variables DataFrame.
X_ts_conditional_hcapm = sm.add_constant(X_ts_conditional_hcapm)

# 5. Iterate through each portfolio excess return column:
print("Performing time-series regressions for Conditional HCAPM (cay as multiplier)...")
for portfolio_col in portfolio_excess_return_cols:
    # a. Define the dependent variable as the current portfolio's excess return.
    y_ts_conditional_hcapm = merged_df[portfolio_col]

    # b. Drop rows with NaN values in the dependent variable and the corresponding rows in the independent variables DataFrame.
    valid_indices = y_ts_conditional_hcapm.dropna().index
    y_ts_cleaned_conditional_hcapm = y_ts_conditional_hcapm.loc[valid_indices]
    X_ts_cleaned_conditional_hcapm = X_ts_conditional_hcapm.loc[valid_indices]

    # Drop rows with NaN values in X_ts_cleaned_conditional_hcapm before fitting the model
    valid_X_indices = X_ts_cleaned_conditional_hcapm.dropna().index
    y_ts_cleaned_conditional_hcapm = y_ts_cleaned_conditional_hcapm.loc[valid_X_indices]
    X_ts_cleaned_conditional_hcapm = X_ts_cleaned_conditional_hcapm.loc[valid_X_indices]


    if not X_ts_cleaned_conditional_hcapm.empty and not y_ts_cleaned_conditional_hcapm.empty:
        # c. Fit an OLS model using sm.OLS with the cleaned dependent and independent variables.
        model_ts_conditional_hcapm = sm.OLS(y_ts_cleaned_conditional_hcapm, X_ts_cleaned_conditional_hcapm)
        results_ts_conditional_hcapm = model_ts_conditional_hcapm.fit()

        # d. Store the estimated coefficients in the dictionary.
        portfolio_betas_conditional_hcapm[portfolio_col] = {
            'beta_market': results_ts_conditional_hcapm.params['market_excess_return'],
            'beta_human_capital': results_ts_conditional_hcapm.params['human_capital_return'],
            'beta_human_capital_x_cay': results_ts_conditional_hcapm.params['human_capital_return_x_cay']
        }
    else:
         # Handle cases where there are no valid data points for regression
        portfolio_betas_conditional_hcapm[portfolio_col] = {
            'beta_market': np.nan,
            'beta_human_capital': np.nan,
            'beta_human_capital_x_cay': np.nan
        }

# 6. Print the estimated betas for each portfolio.
print("Time-series regressions complete. Estimated Betas for Conditional HCAPM:")
for portfolio, betas in portfolio_betas_conditional_hcapm.items():
    print(f"{portfolio}:")
    print(f"  Beta Market: {betas['beta_market']:.4f}")
    print(f"  Beta Human Capital: {betas['beta_human_capital']:.4f}") # Corrected key name
    print(f"  Beta Human Capital x CAY: {betas['beta_human_capital_x_cay']:.4f}")


# 7. Create a DataFrame of the estimated betas for cross-sectional regression and add a constant for lambda_0.
betas_df_conditional_hcapm = pd.DataFrame.from_dict(portfolio_betas_conditional_hcapm, orient='index')
betas_df_conditional_hcapm = sm.add_constant(betas_df_conditional_hcapm) # Add constant for lambda_0

# 8. Perform cross-sectional regressions
print("\nPerforming cross-sectional regressions for Conditional HCAPM...")
lambda_estimates_conditional_hcapm = []

for index, row in merged_df.iterrows():
    y_cs_conditional_hcapm = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs_conditional_hcapm and get the corresponding betas
    valid_portfolios = y_cs_conditional_hcapm.dropna().index
    y_cs_cleaned_conditional_hcapm = y_cs_conditional_hcapm.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs_conditional_hcapm = betas_df_conditional_hcapm.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 4 for intercept and three betas)
    if len(valid_portfolios) >= 4 and not betas_for_cs_conditional_hcapm.empty:
        try:
            model_cs_conditional_hcapm = sm.OLS(y_cs_cleaned_conditional_hcapm, betas_for_cs_conditional_hcapm)
            results_cs_conditional_hcapm = model_cs_conditional_hcapm.fit()
            # Ensure the number of parameters is 4 before appending (lambda_0, lambda_market, lambda_human_capital, lambda_human_capital_x_cay)
            if len(results_cs_conditional_hcapm.params) == 4:
                lambda_estimates_conditional_hcapm.append(results_cs_conditional_hcapm.params.tolist())
            else:
                 # Append NaNs if the number of parameters is not as expected
                lambda_estimates_conditional_hcapm.append([np.nan, np.nan, np.nan, np.nan])
        except Exception as e:
            # Handle potential errors during regression
            print(f"Regression error at index {index}: {e}")
            lambda_estimates_conditional_hcapm.append([np.nan, np.nan, np.nan, np.nan])
    else:
        # Append NaNs if regression cannot be performed
        lambda_estimates_conditional_hcapm.append([np.nan, np.nan, np.nan, np.nan])

lambda_df_conditional_hcapm = pd.DataFrame(lambda_estimates_conditional_hcapm, columns=['lambda_0', 'lambda_market', 'lambda_human_capital', 'lambda_human_capital_x_cay'])
print("Cross-sectional regressions complete.")
#display(lambda_df_conditional_hcapm.head()) # Optional: display head of lambda_df_conditional_hcapm


# 9. Compute Fama–MacBeth lambdas and standard errors
print("\nComputing Fama-MacBeth lambdas and standard errors for Conditional HCAPM...")
lambda_0_mean_conditional_hcapm = lambda_df_conditional_hcapm['lambda_0'].mean()
lambda_0_sem_conditional_hcapm = lambda_df_conditional_hcapm['lambda_0'].sem()
lambda_market_mean_conditional_hcapm = lambda_df_conditional_hcapm['lambda_market'].mean()
lambda_market_sem_conditional_hcapm = lambda_df_conditional_hcapm['lambda_market'].sem()
lambda_human_capital_mean_conditional_hcapm = lambda_df_conditional_hcapm['lambda_human_capital'].mean()
lambda_human_capital_sem_conditional_hcapm = lambda_df_conditional_hcapm['lambda_human_capital'].sem()
lambda_human_capital_x_cay_mean_conditional_hcapm = lambda_df_conditional_hcapm['lambda_human_capital_x_cay'].mean()
lambda_human_capital_x_cay_sem_conditional_hcapm = lambda_df_conditional_hcapm['lambda_human_capital_x_cay'].sem()


print("Fama-MacBeth Lambda Estimates (Conditional HCAPM):")
print(f"Mean of lambda_0: {lambda_0_mean_conditional_hcapm:.4f}")
print(f"Standard Error of lambda_0: {lambda_0_sem_conditional_hcapm:.4f}")
print(f"Mean of lambda_market: {lambda_market_mean_conditional_hcapm:.4f}")
print(f"Standard Error of lambda_market: {lambda_market_sem_conditional_hcapm:.4f}")
print(f"Mean of lambda_human_capital: {lambda_human_capital_mean_conditional_hcapm:.4f}")
print(f"Standard Error of lambda_human_capital: {lambda_human_capital_sem_conditional_hcapm:.4f}")
print(f"Mean of lambda_human_capital_x_cay: {lambda_human_capital_x_cay_mean_conditional_hcapm:.4f}")
print(f"Standard Error of lambda_human_capital_x_cay: {lambda_human_capital_x_cay_sem_conditional_hcapm:.4f}")


# 10. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factors
# Factors are market excess return, human capital return, and human capital return x cay interaction
factors_conditional_hcapm_shanken = merged_df[['market_excess_return', 'human_capital_return', 'human_capital_return_x_cay']].dropna()
mu_f_conditional_hcapm = factors_conditional_hcapm_shanken.mean().values
Sigma_f_conditional_hcapm = factors_conditional_hcapm_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
try:
    Sigma_f_inv_conditional_hcapm = np.linalg.inv(Sigma_f_conditional_hcapm)
    shanken_correction_factor_conditional_hcapm = (1 + mu_f_conditional_hcapm.T @ Sigma_f_inv_conditional_hcapm @ mu_f_conditional_hcapm)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_conditional_hcapm = np.nan


if not np.isnan(shanken_correction_factor_conditional_hcapm):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_conditional_hcapm = (lambda_0_sem_conditional_hcapm**2) * shanken_correction_factor_conditional_hcapm
    lambda_market_shanken_variance_conditional_hcapm = (lambda_market_sem_conditional_hcapm**2) * shanken_correction_factor_conditional_hcapm
    lambda_human_capital_shanken_variance_conditional_hcapm = (lambda_human_capital_sem_conditional_hcapm**2) * shanken_correction_factor_conditional_hcapm
    lambda_human_capital_x_cay_shanken_variance_conditional_hcapm = (lambda_human_capital_x_cay_sem_conditional_hcapm**2) * shanken_correction_factor_conditional_hcapm


    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_conditional_hcapm = np.sqrt(lambda_0_shanken_variance_conditional_hcapm)
    lambda_market_shanken_sem_conditional_hcapm = np.sqrt(lambda_market_shanken_variance_conditional_hcapm)
    lambda_human_capital_shanken_sem_conditional_hcapm = np.sqrt(lambda_human_capital_shanken_variance_conditional_hcapm)
    lambda_human_capital_x_cay_shanken_sem_conditional_hcapm = np.sqrt(lambda_human_capital_x_cay_shanken_variance_conditional_hcapm)


    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_conditional_hcapm = lambda_0_mean_conditional_hcapm / lambda_0_shanken_sem_conditional_hcapm
    lambda_market_shanken_tstat_conditional_hcapm = lambda_market_mean_conditional_hcapm / lambda_market_shanken_sem_conditional_hcapm
    lambda_human_capital_shanken_tstat_conditional_hcapm = lambda_human_capital_mean_conditional_hcapm / lambda_human_capital_shanken_sem_conditional_hcapm
    lambda_human_capital_x_cay_shanken_tstat_conditional_hcapm = lambda_human_capital_x_cay_mean_conditional_hcapm / lambda_human_capital_x_cay_shanken_sem_conditional_hcapm


    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (Conditional HCAPM):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_conditional_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_conditional_hcapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_market: {lambda_market_shanken_sem_conditional_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat_conditional_hcapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_human_capital: {lambda_human_capital_shanken_sem_conditional_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_human_capital: {lambda_human_capital_shanken_tstat_conditional_hcapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_human_capital_x_cay: {lambda_human_capital_x_cay_shanken_sem_conditional_hcapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_human_capital_x_cay: {lambda_human_capital_x_cay_shanken_tstat_conditional_hcapm:.4f}")

else:
    print("\nShanken correction could not be applied.")


# 11. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_conditional_hcapm = []

for index, row in merged_df.iterrows():
    y_cs_conditional_hcapm = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs_conditional_hcapm and get the corresponding betas
    valid_portfolios = y_cs_conditional_hcapm.dropna().index
    y_cs_cleaned_conditional_hcapm = y_cs_conditional_hcapm.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs_conditional_hcapm = betas_df_conditional_hcapm.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 4 for intercept and three betas)
    if len(valid_portfolios) >= 4 and not betas_for_cs_conditional_hcapm.empty:
        try:
            model_cs_conditional_hcapm = sm.OLS(y_cs_cleaned_conditional_hcapm, betas_for_cs_conditional_hcapm)
            results_cs_conditional_hcapm = model_cs_conditional_hcapm.fit()
            if len(results_cs_conditional_hcapm.params) == 4:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned_conditional_hcapm.mean()
                ss_total = ((y_cs_cleaned_conditional_hcapm - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs_conditional_hcapm.predict(betas_for_cs_conditional_hcapm)
                ss_residual = ((y_cs_cleaned_conditional_hcapm - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_conditional_hcapm.append(cs_r_squared)
                else:
                    cs_r_squared_values_conditional_hcapm.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_conditional_hcapm.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_conditional_hcapm.append(np.nan)
    else:
        cs_r_squared_values_conditional_hcapm.append(np.nan)


average_cs_r_squared_conditional_hcapm = np.nanmean(cs_r_squared_values_conditional_hcapm)

print(f"Average Cross-Sectional R-squared (Conditional HCAPM): {average_cs_r_squared_conditional_hcapm:.4f}")


# 12. Report results (Summary)
print("\n--- Conditional HCAPM Model Results (cay as multiplier) ---")
print(f"Mean lambda_market: {lambda_market_mean_conditional_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_market: {lambda_market_shanken_tstat_conditional_hcapm:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_market: N/A (Shanken correction not applied)")

print(f"Mean lambda_human_capital: {lambda_human_capital_mean_conditional_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_human_capital: {lambda_human_capital_shanken_tstat_conditional_hcapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_human_capital: N/A (Shanken correction not applied)")

print(f"Mean lambda_human_capital_x_cay: {lambda_human_capital_x_cay_mean_conditional_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_human_capital_x_cay: {lambda_human_capital_x_cay_shanken_tstat_conditional_hcapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_human_capital_x_cay: N/A (Shanken correction not applied)")

print(f"Mean lambda_0: {lambda_0_mean_conditional_hcapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_hcapm):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_conditional_hcapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")

print(f"Average cross-sectional R-squared: {average_cs_r_squared_conditional_hcapm:.4f}")

print("\n--- Conditional HCAPM Summary ---")
print("Key Findings:")
print(f"* The estimated betas for the six portfolios show varying sensitivities to market risk, human capital risk, and conditional human capital risk (human capital x cay).")
print(f"* The mean estimated market risk premium (lambda_market) is {lambda_market_mean_conditional_hcapm:.4f}.")
print(f"* The mean estimated human capital risk premium (lambda_human_capital) is {lambda_human_capital_mean_conditional_hcapm:.4f}.")
print(f"* The mean estimated conditional human capital risk premium (lambda_human_capital_x_cay) is {lambda_human_capital_x_cay_mean_conditional_hcapm:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_conditional_hcapm:.4f}.")

if not np.isnan(shanken_correction_factor_conditional_hcapm):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - Market Risk Premium t-statistic: {lambda_market_shanken_tstat_conditional_hcapm:.4f}")
    print(f"  - Human Capital Risk Premium t-statistic: {lambda_human_capital_shanken_tstat_conditional_hcapm:.4f}")
    print(f"  - Conditional Human Capital Risk Premium t-statistic: {lambda_human_capital_x_cay_shanken_tstat_conditional_hcapm:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_conditional_hcapm:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_conditional_hcapm:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Estimate the Conditional CCAPM model (cay as multiplier).")
print("- Summarize all conditional model results and compare them to the unconditional models.")
```

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

# 1. Define a list containing the names of the portfolio excess return columns.
portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

# 2. Create a dictionary to store the estimated betas for each portfolio (Conditional CCAPM).
portfolio_betas_conditional_ccapm = {}

# 3. Define the independent variables for the time-series regression as 'consumption_growth' and 'consumption_growth_x_cay'.
X_ts_conditional_ccapm = merged_df[['consumption_growth', 'consumption_growth_x_cay']].copy()

# 4. Add a constant term to the independent variables DataFrame.
X_ts_conditional_ccapm = sm.add_constant(X_ts_conditional_ccapm)

# 5. Iterate through each portfolio excess return column:
print("Performing time-series regressions for Conditional CCAPM (cay as multiplier)...")
for portfolio_col in portfolio_excess_return_cols:
    # a. Define the dependent variable as the current portfolio's excess return.
    y_ts_conditional_ccapm = merged_df[portfolio_col]

    # b. Drop rows with NaN values in the dependent variable and the corresponding rows in the independent variables DataFrame.
    valid_indices = y_ts_conditional_ccapm.dropna().index
    y_ts_cleaned_conditional_ccapm = y_ts_conditional_ccapm.loc[valid_indices]
    X_ts_cleaned_conditional_ccapm = X_ts_conditional_ccapm.loc[valid_indices]

    # Drop rows with NaN values in X_ts_cleaned_conditional_ccapm before fitting the model
    valid_X_indices = X_ts_cleaned_conditional_ccapm.dropna().index
    y_ts_cleaned_conditional_ccapm = y_ts_cleaned_conditional_ccapm.loc[valid_X_indices]
    X_ts_cleaned_conditional_ccapm = X_ts_cleaned_conditional_ccapm.loc[valid_X_indices]


    if not X_ts_cleaned_conditional_ccapm.empty and not y_ts_cleaned_conditional_ccapm.empty:
        # c. Fit an OLS model using sm.OLS with the cleaned dependent and independent variables.
        model_ts_conditional_ccapm = sm.OLS(y_ts_cleaned_conditional_ccapm, X_ts_cleaned_conditional_ccapm)
        results_ts_conditional_ccapm = model_ts_conditional_ccapm.fit()

        # d. Store the estimated coefficients in the dictionary.
        portfolio_betas_conditional_ccapm[portfolio_col] = {
            'beta_consumption': results_ts_conditional_ccapm.params['consumption_growth'],
            'beta_consumption_x_cay': results_ts_conditional_ccapm.params['consumption_growth_x_cay']
        }
    else:
         # Handle cases where there are no valid data points for regression
        portfolio_betas_conditional_ccapm[portfolio_col] = {
            'beta_consumption': np.nan,
            'beta_consumption_x_cay': np.nan
        }

# 6. Print the estimated betas for each portfolio.
print("Time-series regressions complete. Estimated Betas for Conditional CCAPM:")
for portfolio, betas in portfolio_betas_conditional_ccapm.items():
    print(f"{portfolio}:")
    print(f"  Beta Consumption: {betas['beta_consumption']:.4f}")
    print(f"  Beta Consumption x CAY: {betas['beta_consumption_x_cay']:.4f}")


# 7. Create a DataFrame of the estimated betas for cross-sectional regression and add a constant for lambda_0.
betas_df_conditional_ccapm = pd.DataFrame.from_dict(portfolio_betas_conditional_ccapm, orient='index')
betas_df_conditional_ccapm = sm.add_constant(betas_df_conditional_ccapm) # Add constant for lambda_0

# 8. Perform cross-sectional regressions
print("\nPerforming cross-sectional regressions for Conditional CCAPM...")
lambda_estimates_conditional_ccapm = []

for index, row in merged_df.iterrows():
    y_cs_conditional_ccapm = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs_conditional_ccapm and get the corresponding betas
    valid_portfolios = y_cs_conditional_ccapm.dropna().index
    y_cs_cleaned_conditional_ccapm = y_cs_conditional_ccapm.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs_conditional_ccapm = betas_df_conditional_ccapm.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 3 for intercept and two betas)
    if len(valid_portfolios) >= 3 and not betas_for_cs_conditional_ccapm.empty:
        try:
            model_cs_conditional_ccapm = sm.OLS(y_cs_cleaned_conditional_ccapm, betas_for_cs_conditional_ccapm)
            results_cs_conditional_ccapm = model_cs_conditional_ccapm.fit()
            # Ensure the number of parameters is 3 before appending (lambda_0, lambda_consumption, lambda_consumption_x_cay)
            if len(results_cs_conditional_ccapm.params) == 3:
                lambda_estimates_conditional_ccapm.append(results_cs_conditional_ccapm.params.tolist())
            else:
                 # Append NaNs if the number of parameters is not as expected
                lambda_estimates_conditional_ccapm.append([np.nan, np.nan, np.nan])
        except Exception as e:
            # Handle potential errors during regression
            print(f"Regression error at index {index}: {e}")
            lambda_estimates_conditional_ccapm.append([np.nan, np.nan, np.nan])
    else:
        # Append NaNs if regression cannot be performed
        lambda_estimates_conditional_ccapm.append([np.nan, np.nan, np.nan])

lambda_df_conditional_ccapm = pd.DataFrame(lambda_estimates_conditional_ccapm, columns=['lambda_0', 'lambda_consumption', 'lambda_consumption_x_cay'])
print("Cross-sectional regressions complete.")
#display(lambda_df_conditional_ccapm.head()) # Optional: display head of lambda_df_conditional_ccapm


# 9. Compute Fama–MacBeth lambdas and standard errors
print("\nComputing Fama-MacBeth lambdas and standard errors for Conditional CCAPM...")
lambda_0_mean_conditional_ccapm = lambda_df_conditional_ccapm['lambda_0'].mean()
lambda_0_sem_conditional_ccapm = lambda_df_conditional_ccapm['lambda_0'].sem()
lambda_consumption_mean_conditional_ccapm = lambda_df_conditional_ccapm['lambda_consumption'].mean()
lambda_consumption_sem_conditional_ccapm = lambda_df_conditional_ccapm['lambda_consumption'].sem()
lambda_consumption_x_cay_mean_conditional_ccapm = lambda_df_conditional_ccapm['lambda_consumption_x_cay'].mean()
lambda_consumption_x_cay_sem_conditional_ccapm = lambda_df_conditional_ccapm['lambda_consumption_x_cay'].sem()


print("Fama-MacBeth Lambda Estimates (Conditional CCAPM):")
print(f"Mean of lambda_0: {lambda_0_mean_conditional_ccapm:.4f}")
print(f"Standard Error of lambda_0: {lambda_0_sem_conditional_ccapm:.4f}")
print(f"Mean of lambda_consumption: {lambda_consumption_mean_conditional_ccapm:.4f}")
print(f"Standard Error of lambda_consumption: {lambda_consumption_sem_conditional_ccapm:.4f}")
print(f"Mean of lambda_consumption_x_cay: {lambda_consumption_x_cay_mean_conditional_ccapm:.4f}")
print(f"Standard Error of lambda_consumption_x_cay: {lambda_consumption_x_cay_sem_conditional_ccapm:.4f}")


# 10. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factors
# Factors are consumption growth and consumption growth x cay interaction
factors_conditional_ccapm_shanken = merged_df[['consumption_growth', 'consumption_growth_x_cay']].dropna()
mu_f_conditional_ccapm = factors_conditional_ccapm_shanken.mean().values
Sigma_f_conditional_ccapm = factors_conditional_ccapm_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
try:
    Sigma_f_inv_conditional_ccapm = np.linalg.inv(Sigma_f_conditional_ccapm)
    shanken_correction_factor_conditional_ccapm = (1 + mu_f_conditional_ccapm.T @ Sigma_f_inv_conditional_ccapm @ mu_f_conditional_ccapm)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_conditional_ccapm = np.nan


if not np.isnan(shanken_correction_factor_conditional_ccapm):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_conditional_ccapm = (lambda_0_sem_conditional_ccapm**2) * shanken_correction_factor_conditional_ccapm
    lambda_consumption_shanken_variance_conditional_ccapm = (lambda_consumption_sem_conditional_ccapm**2) * shanken_correction_factor_conditional_ccapm
    lambda_consumption_x_cay_shanken_variance_conditional_ccapm = (lambda_consumption_x_cay_sem_conditional_ccapm**2) * shanken_correction_factor_conditional_ccapm


    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_conditional_ccapm = np.sqrt(lambda_0_shanken_variance_conditional_ccapm)
    lambda_consumption_shanken_sem_conditional_ccapm = np.sqrt(lambda_consumption_shanken_variance_conditional_ccapm)
    lambda_consumption_x_cay_shanken_sem_conditional_ccapm = np.sqrt(lambda_consumption_x_cay_shanken_variance_conditional_ccapm)


    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_conditional_ccapm = lambda_0_mean_conditional_ccapm / lambda_0_shanken_sem_conditional_ccapm
    lambda_consumption_shanken_tstat_conditional_ccapm = lambda_consumption_mean_conditional_ccapm / lambda_consumption_shanken_sem_conditional_ccapm
    lambda_consumption_x_cay_shanken_tstat_conditional_ccapm = lambda_consumption_x_cay_mean_conditional_ccapm / lambda_consumption_x_cay_shanken_sem_conditional_ccapm


    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (Conditional CCAPM):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_conditional_ccapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_conditional_ccapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_consumption: {lambda_consumption_shanken_sem_conditional_ccapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_consumption: {lambda_consumption_shanken_tstat_conditional_ccapm:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_consumption_x_cay: {lambda_consumption_x_cay_shanken_sem_conditional_ccapm:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_consumption_x_cay: {lambda_consumption_x_cay_shanken_tstat_conditional_ccapm:.4f}")

else:
    print("\nShanken correction could not be applied.")


# 11. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_conditional_ccapm = []

for index, row in merged_df.iterrows():
    y_cs_conditional_ccapm = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs_conditional_ccapm and get the corresponding betas
    valid_portfolios = y_cs_conditional_ccapm.dropna().index
    y_cs_cleaned_conditional_ccapm = y_cs_conditional_ccapm.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs_conditional_ccapm = betas_df_conditional_ccapm.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 3 for intercept and two betas)
    if len(valid_portfolios) >= 3 and not betas_for_cs_conditional_ccapm.empty:
        try:
            model_cs_conditional_ccapm = sm.OLS(y_cs_cleaned_conditional_ccapm, betas_for_cs_conditional_ccapm)
            results_cs_conditional_ccapm = model_cs_conditional_ccapm.fit()
            if len(results_cs_conditional_ccapm.params) == 3:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned_conditional_ccapm.mean()
                ss_total = ((y_cs_cleaned_conditional_ccapm - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs_conditional_ccapm.predict(betas_for_cs_conditional_ccapm)
                ss_residual = ((y_cs_cleaned_conditional_ccapm - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_conditional_ccapm.append(cs_r_squared)
                else:
                    cs_r_squared_values_conditional_ccapm.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_conditional_ccapm.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_conditional_ccapm.append(np.nan)
    else:
        cs_r_squared_values_conditional_ccapm.append(np.nan)


average_cs_r_squared_conditional_ccapm = np.nanmean(cs_r_squared_values_conditional_ccapm)

print(f"Average Cross-Sectional R-squared (Conditional CCAPM): {average_cs_r_squared_conditional_ccapm:.4f}")


# 12. Report results (Summary)
print("\n--- Conditional CCAPM Model Results (cay as multiplier) ---")
print(f"Mean lambda_consumption: {lambda_consumption_mean_conditional_ccapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_ccapm):
    print(f"Shanken-corrected t-statistic for lambda_consumption: {lambda_consumption_shanken_tstat_conditional_ccapm:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_consumption: N/A (Shanken correction not applied)")

print(f"Mean lambda_consumption_x_cay: {lambda_consumption_x_cay_mean_conditional_ccapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_ccapm):
    print(f"Shanken-corrected t-statistic for lambda_consumption_x_cay: {lambda_consumption_x_cay_shanken_tstat_conditional_ccapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_consumption_x_cay: N/A (Shanken correction not applied)")

print(f"Mean lambda_0: {lambda_0_mean_conditional_ccapm:.4f}")
if not np.isnan(shanken_correction_factor_conditional_ccapm):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_conditional_ccapm:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")

print(f"Average cross-sectional R-squared: {average_cs_r_squared_conditional_ccapm:.4f}")

print("\n--- Conditional CCAPM Summary ---")
print("Key Findings:")
print(f"* The estimated betas for the six portfolios show varying sensitivities to consumption growth and conditional consumption growth (consumption growth x cay).")
print(f"* The mean estimated consumption risk premium (lambda_consumption) is {lambda_consumption_mean_conditional_ccapm:.4f}.")
print(f"* The mean estimated conditional consumption risk premium (lambda_consumption_x_cay) is {lambda_consumption_x_cay_mean_conditional_ccapm:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_conditional_ccapm:.4f}.")

if not np.isnan(shanken_correction_factor_conditional_ccapm):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - Consumption Risk Premium t-statistic: {lambda_consumption_shanken_tstat_conditional_ccapm:.4f}")
    print(f"  - Conditional Consumption Risk Premium t-statistic: {lambda_consumption_x_cay_shanken_tstat_conditional_ccapm:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_conditional_ccapm:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_conditional_ccapm:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Summarize all conditional model results and compare them to the unconditional models.")
```

The next step is to perform the cross-sectional regressions, compute the Fama-MacBeth lambdas and standard errors, apply the Shanken correction, and compute the average cross-sectional R² as per the overall task instructions.
```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# 7. Create a DataFrame of the estimated betas for cross-sectional regression and add a constant for lambda_0.
betas_df_ff3 = pd.DataFrame.from_dict(portfolio_betas_ff3, orient='index')
betas_df_ff3 = sm.add_constant(betas_df_ff3) # Add constant for lambda_0

# 8. Perform cross-sectional regressions
print("\nPerforming cross-sectional regressions for Fama-French 3-Factor Model...")
lambda_estimates_ff3 = []

portfolio_excess_return_cols = [
    'BV_excess_return', 'BN_excess_return', 'BG_excess_return',
    'SV_excess_return', 'SN_excess_return', 'SG_excess_return'
]

for index, row in merged_df.iterrows():
    y_cs_ff3 = row[portfolio_excess_return_cols]

    # Drop NaN values from y_cs_ff3 and get the corresponding betas from betas_df_ff3
    valid_portfolios = y_cs_ff3.dropna().index
    y_cs_cleaned_ff3 = y_cs_ff3.loc[valid_portfolios].astype(float)

    # Select the betas corresponding to the valid portfolios
    betas_for_cs_ff3 = betas_df_ff3.loc[valid_portfolios].astype(float)

    # Ensure there are enough valid data points and betas for regression (at least 4 for intercept and three betas)
    if len(valid_portfolios) >= 4 and not betas_for_cs_ff3.empty:
        try:
            model_cs_ff3 = sm.OLS(y_cs_cleaned_ff3, betas_for_cs_ff3)
            results_cs_ff3 = model_cs_ff3.fit()
            # Ensure the number of parameters is 4 before appending (lambda_0, lambda_smb, lambda_hml, lambda_wml)
            if len(results_cs_ff3.params) == 4:
                lambda_estimates_ff3.append(results_cs_ff3.params.tolist())
            else:
                 # Append NaNs if the number of parameters is not as expected
                lambda_estimates_ff3.append([np.nan, np.nan, np.nan, np.nan])
        except Exception as e:
            # Handle potential errors during regression
            print(f"Regression error at index {index}: {e}")
            lambda_estimates_ff3.append([np.nan, np.nan, np.nan, np.nan])
    else:
        # Append NaNs if regression cannot be performed
        lambda_estimates_ff3.append([np.nan, np.nan, np.nan, np.nan])

lambda_df_ff3 = pd.DataFrame(lambda_estimates_ff3, columns=['lambda_0', 'lambda_smb', 'lambda_hml', 'lambda_wml'])
print("Cross-sectional regressions complete.")
#display(lambda_df_ff3.head()) # Optional: display head of lambda_df_ff3


# 9. Compute Fama–MacBeth lambdas and standard errors
print("\nComputing Fama-MacBeth lambdas and standard errors for Fama-French 3-Factor Model...")
lambda_0_mean_ff3 = lambda_df_ff3['lambda_0'].mean()
lambda_0_sem_ff3 = lambda_df_ff3['lambda_0'].sem()
lambda_smb_mean_ff3 = lambda_df_ff3['lambda_smb'].mean()
lambda_smb_sem_ff3 = lambda_df_ff3['lambda_smb'].sem()
lambda_hml_mean_ff3 = lambda_df_ff3['lambda_hml'].mean()
lambda_hml_sem_ff3 = lambda_df_ff3['lambda_hml'].sem()
lambda_wml_mean_ff3 = lambda_df_ff3['lambda_wml'].mean()
lambda_wml_sem_ff3 = lambda_df_ff3['lambda_wml'].sem()


print("Fama-MacBeth Lambda Estimates (Fama-French 3-Factor Model):")
print(f"Mean of lambda_0: {lambda_0_mean_ff3:.4f}")
print(f"Standard Error of lambda_0: {lambda_0_sem_ff3:.4f}")
print(f"Mean of lambda_smb: {lambda_smb_mean_ff3:.4f}")
print(f"Standard Error of lambda_smb: {lambda_smb_sem_ff3:.4f}")
print(f"Mean of lambda_hml: {lambda_hml_mean_ff3:.4f}")
print(f"Standard Error of lambda_hml: {lambda_hml_sem_ff3:.4f}")
print(f"Mean of lambda_wml: {lambda_wml_mean_ff3:.4f}")
print(f"Standard Error of lambda_wml: {lambda_wml_sem_ff3:.4f}")


# 10. Apply Shanken correction
print("\nApplying Shanken correction...")
# 1. Calculate the mean vector and covariance matrix of the factor returns
# Factors are SMB, HML, and WML
factors_ff3_shanken = merged_df[['smb', 'hml', 'wml']].dropna()
mu_f_ff3 = factors_ff3_shanken.mean().values
Sigma_f_ff3 = factors_ff3_shanken.cov().values

# 2. Calculate the Shanken correction factor
# Formula: (1 + mu_f' * inv(Sigma_f) * mu_f)
try:
    Sigma_f_inv_ff3 = np.linalg.inv(Sigma_f_ff3)
    shanken_correction_factor_ff3 = (1 + mu_f_ff3.T @ Sigma_f_inv_ff3 @ mu_f_ff3)
except np.linalg.LinAlgError:
    print("Could not compute inverse of the covariance matrix. Skipping Shanken correction.")
    shanken_correction_factor_ff3 = np.nan


if not np.isnan(shanken_correction_factor_ff3):
    # 3. Multiply the squared standard errors of the Fama-MacBeth lambdas by the Shanken correction factor
    lambda_0_shanken_variance_ff3 = (lambda_0_sem_ff3**2) * shanken_correction_factor_ff3
    lambda_smb_shanken_variance_ff3 = (lambda_smb_sem_ff3**2) * shanken_correction_factor_ff3
    lambda_hml_shanken_variance_ff3 = (lambda_hml_sem_ff3**2) * shanken_correction_factor_ff3
    lambda_wml_shanken_variance_ff3 = (lambda_wml_sem_ff3**2) * shanken_correction_factor_ff3


    # 4. Take the square root of the Shanken-corrected variances to get the Shanken-corrected standard errors
    lambda_0_shanken_sem_ff3 = np.sqrt(lambda_0_shanken_variance_ff3)
    lambda_smb_shanken_sem_ff3 = np.sqrt(lambda_smb_shanken_variance_ff3)
    lambda_hml_shanken_sem_ff3 = np.sqrt(lambda_hml_shanken_variance_ff3)
    lambda_wml_shanken_sem_ff3 = np.sqrt(lambda_wml_shanken_variance_ff3)


    # 5. Calculate the t-statistics using the mean lambdas and Shanken-corrected standard errors
    lambda_0_shanken_tstat_ff3 = lambda_0_mean_ff3 / lambda_0_shanken_sem_ff3
    lambda_smb_shanken_tstat_ff3 = lambda_smb_mean_ff3 / lambda_smb_shanken_sem_ff3
    lambda_hml_shanken_tstat_ff3 = lambda_hml_mean_ff3 / lambda_hml_shanken_sem_ff3
    lambda_wml_shanken_tstat_ff3 = lambda_wml_mean_ff3 / lambda_wml_shanken_sem_ff3


    # 6. Print the Shanken-corrected standard errors and t-statistics
    print("\nShanken-corrected Results (Fama-French 3-Factor Model):")
    print(f"Shanken-corrected Standard Error of lambda_0: {lambda_0_shanken_sem_ff3:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_ff3:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_smb: {lambda_smb_shanken_sem_ff3:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_smb: {lambda_smb_shanken_tstat_ff3:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_hml: {lambda_hml_shanken_sem_ff3:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_hml: {lambda_hml_shanken_tstat_ff3:.4f}")
    print(f"Shanken-corrected Standard Error of lambda_wml: {lambda_wml_shanken_sem_ff3:.4f}")
    print(f"Shanken-corrected t-statistic for lambda_wml: {lambda_wml_shanken_tstat_ff3:.4f}")

else:
    print("\nShanken correction could not be applied.")


# 11. Compute cross-sectional R²
print("\nComputing average cross-sectional R-squared...")
cs_r_squared_values_ff3 = []

for index, row in merged_df.iterrows():
    y_cs_ff3 = row[portfolio_excess_return_cols]

    valid_portfolios = y_cs_ff3.dropna().index
    y_cs_cleaned_ff3 = y_cs_ff3.loc[valid_portfolios].astype(float)

    betas_for_cs_ff3 = betas_df_ff3.loc[valid_portfolios].astype(float)

    if len(valid_portfolios) >= 4 and not betas_for_cs_ff3.empty:
        try:
            model_cs_ff3 = sm.OLS(y_cs_cleaned_ff3, betas_for_cs_ff3)
            results_cs_ff3 = model_cs_ff3.fit()
            if len(results_cs_ff3.params) == 4:
                 # Calculate R-squared for the cross-sectional regression
                y_cs_mean = y_cs_cleaned_ff3.mean()
                ss_total = ((y_cs_cleaned_ff3 - y_cs_mean) ** 2).sum()
                # Calculate predicted values using the cross-sectional regression results
                predicted_y_cs = results_cs_ff3.predict(betas_for_cs_ff3)
                ss_residual = ((y_cs_cleaned_ff3 - predicted_y_cs) ** 2).sum()
                # Avoid division by zero if ss_total is zero
                if ss_total > 0:
                    cs_r_squared = 1 - (ss_residual / ss_total)
                    cs_r_squared_values_ff3.append(cs_r_squared)
                else:
                    cs_r_squared_values_ff3.append(np.nan) # Append NaN if ss_total is zero
            else:
                cs_r_squared_values_ff3.append(np.nan)
        except Exception as e:
            print(f"Regression error at index {index}: {e}")
            cs_r_squared_values_ff3.append(np.nan)
    else:
        cs_r_squared_values_ff3.append(np.nan)


average_cs_r_squared_ff3 = np.nanmean(cs_r_squared_values_ff3)

print(f"Average Cross-Sectional R-squared (Fama-French 3-Factor Model): {average_cs_r_squared_ff3:.4f}")


# 12. Report results (Summary)
print("\n--- Fama-French 3-Factor Model Results ---")
print(f"Mean lambda_smb: {lambda_smb_mean_ff3:.4f}")
if not np.isnan(shanken_correction_factor_ff3):
    print(f"Shanken-corrected t-statistic for lambda_smb: {lambda_smb_shanken_tstat_ff3:.4f}")
else:
     print("Shanken-corrected t-statistic for lambda_smb: N/A (Shanken correction not applied)")

print(f"Mean lambda_hml: {lambda_hml_mean_ff3:.4f}")
if not np.isnan(shanken_correction_factor_ff3):
    print(f"Shanken-corrected t-statistic for lambda_hml: {lambda_hml_shanken_tstat_ff3:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_hml: N/A (Shanken correction not applied)")

print(f"Mean lambda_wml: {lambda_wml_mean_ff3:.4f}")
if not np.isnan(shanken_correction_factor_ff3):
    print(f"Shanken-corrected t-statistic for lambda_wml: {lambda_wml_shanken_tstat_ff3:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_wml: N/A (Shanken correction not applied)")


print(f"Mean lambda_0: {lambda_0_mean_ff3:.4f}")
if not np.isnan(shanken_correction_factor_ff3):
    print(f"Shanken-corrected t-statistic for lambda_0: {lambda_0_shanken_tstat_ff3:.4f}")
else:
    print("Shanken-corrected t-statistic for lambda_0: N/A (Shanken correction not applied)")

print(f"Average cross-sectional R-squared: {average_cs_r_squared_ff3:.4f}")

print("\n--- Fama-French 3-Factor Model Summary ---")
print("Key Findings:")
print(f"* The estimated betas for the six portfolios show varying sensitivities to SMB, HML, and WML factors.")
print(f"* The mean estimated SMB risk premium (lambda_smb) is {lambda_smb_mean_ff3:.4f}.")
print(f"* The mean estimated HML risk premium (lambda_hml) is {lambda_hml_mean_ff3:.4f}.")
print(f"* The mean estimated WML risk premium (lambda_wml) is {lambda_wml_mean_ff3:.4f}.")
print(f"* The mean estimated zero-beta rate (lambda_0) is {lambda_0_mean_ff3:.4f}.")

if not np.isnan(shanken_correction_factor_ff3):
    print("* Shanken-corrected t-statistics indicate the statistical significance of the estimated risk premia.")
    print(f"  - SMB Risk Premium t-statistic: {lambda_smb_shanken_tstat_ff3:.4f}")
    print(f"  - HML Risk Premium t-statistic: {lambda_hml_shanken_tstat_ff3:.4f}")
    print(f"  - WML Risk Premium t-statistic: {lambda_wml_shanken_tstat_ff3:.4f}")
    print(f"  - Zero-Beta Rate t-statistic: {lambda_0_shanken_tstat_ff3:.4f}")
else:
    print("* Shanken correction could not be applied due to issues with the covariance matrix.")

print(f"* The average cross-sectional R-squared of {average_cs_r_squared_ff3:.4f} suggests the proportion of the cross-sectional variation in portfolio excess returns explained by the model.")

print("\nNext Steps:")
print("- Summarize all model results in a comprehensive table.")
```
